# performed 3 cluster with pca with hirarical clustring

```python
WCSS=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)# random state 0
    kmeans.fit(sc_x1)
    WCSS.append(kmeans.inertia_)
    print(i,WCSS)
```

```python
plt.plot(range(1,11),WCSS)
plt.title('ELBOW METHOD')
plt.xlabel('NUMBER OF CLUSTERS')
```
